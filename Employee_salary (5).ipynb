# ***LabelEncoder***

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data['Gender']=encoder.fit_transform(data['Gender'])
data['Department']=encoder.fit_transform(data['Department'])
data['Job Title']=encoder.fit_transform(data['Job Title'])
data['Education Level']=encoder.fit_transform(data['Education Level'])
data['Workclass']=encoder.fit_transform(data['Workclass'])
```

```python
# Separating the dependent and independent variables
# data.dropna(subset=['salary_category'], inplace=True)
x = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
# Checking all the Machine Learning Algorithms to create Model

from sklearn.pipeline import Pipeline

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

from sklearn.preprocessing import StandardScaler, OneHotEncoder

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

models = {
    'Logistic Regression': LogisticRegression(),
    'Support Vector Machine': SVC(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Gaussian Naive Bayes': GaussianNB()
}

results = []

for name, model in models.items():
    pipe = Pipeline(steps=[('scaler', StandardScaler()), ('model', model)])
    pipe.fit(x_train, y_train)
    y_pred = pipe.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    results.append((name, acc))


for name, acc in results:
    print(f"{name}: {acc:.4f}")
    print(classification_report(y_test, y_pred))
```

```python
# Comparing the accuracy of the Machine Learning Algorithms with bar graph
import matplotlib.pyplot as plt
import numpy as np

names = [name for name, _ in results]
accuracies = [acc for _, acc in results]

plt.bar(names, accuracies)
plt.xlabel('Machine Learning Algorithms')
plt.ylabel('Accuracy')
plt.title('Comparison of Machine Learning Algorithms')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
%%writefile app.py
import streamlit as sd
import pandas as pd
import joblib

# Load the best model
best_model = joblib.load('best_model.pkl')

# Streamlit app
sd.title("Employee Salary Prediction App")

# Side bar for user input
sd.sidebar.header("User Input")
sd.markdown("Enter the details of the employee to predict their salary scale.")

# Input features
age = sd.sidebar.number_input("Age", min_value=18, max_value=100, value=25)
gender = sd.sidebar.selectbox("Gender", ["Male", "Female", "Other"])
department = sd.sidebar.selectbox("Department", ["Engineering", "IT", "HR", "Finance", "Marketing", "Sales", "Support","Operations"])
education_level = sd.sidebar.selectbox("Education Level", ["Associate Degree", "Bachelor's Degree", "High School", "PhD", "Master's Degree"])
hours_per_week = sd.sidebar.number_input("Hours per Week", min_value=0, max_value=168, value=40)
job_title = sd.sidebar.selectbox("Job Title", [
     "Software Engineer","Accountant ","HR Assistant","Marketing Executive ","Systems Engineer","Financial Analyst",
     "Sales Executive ","Operations Manager ","Sales Manager","IT Support","Customer Support","Tech Support"
     "HR Manager","Network Admin ","Logistics Coordinator","Content Strategist "

])
experience = sd.sidebar.number_input("Years of Experience", min_value=0, max_value=30, value=5)
Workclass = sd.sidebar.selectbox("Workclass", ["Private", "Self-Employed", "Government"])

# Clean and map the input
input_data = pd.DataFrame({
    'Age': [age],
    'Gender': [gender.strip()],
    'Department': [department.strip()],
    'Education Level': [education_level.strip()],
    'Hours per Week': [hours_per_week],
    'Job Title': [job_title.strip()],
    'Years of Experience': [experience],
    'Workclass': [Workclass.strip()]
})

sd.subheader("🔎 User Input Data")
sd.write(input_data)

# Prediction Button
if sd.button("Predict Salary"):
    # Mapping dictionaries

    input_data['Gender'] = input_data['Gender'].map({"Male": 0, "Female": 1, "Other": 2})
    input_data['Department'] = input_data['Department'].map({
        "Engineering": 0, "IT": 1, "HR": 2, "Finance": 3, "Marketing": 4, "Sales": 5, "Support": 6, "Operations": 7
    })
    input_data['Education Level'] = input_data['Education Level'].map({
        "Associate Degree": 0, "Bachelor's Degree": 1, "High School": 2,
        "PhD": 3, "Master's Degree": 4
    })
    input_data['Job Title'] = input_data['Job Title'].map({
      "Software Engineer": 0, "Accountant": 1, "HR Assistant": 2, "Marketing Executive": 3, "Systems Engineer": 4,
      "Financial Analyst": 5, "Sales Executive": 6, "Operations Manager": 7, "Sales Manager": 8, "IT Support": 9
      , "Customer Support": 10, "Tech Support": 11, "HR Manager": 12, "Network Admin": 13, "Logistics Coordinator": 14
      , "Content Strategist": 15
    })
    input_data['Workclass'] = input_data['Workclass'].map({"Private": 0, "Self-Employed": 1, "Government": 2})


    #  Make predictions
    predictions = best_model.predict(input_data)

    # Display the prediction result
    sd.subheader("Predicted Total Salary")
    sd.write(predictions[0])

# Batch prediction section
sd.subheader("Batch Prediction")
uploaded_file = sd.file_uploader("Upload CSV file for batch prediction", type=["csv"])
if uploaded_file is not None:
    batch_data = pd.read_csv(uploaded_file)
    sd.write("Batch Data:")
    sd.write(batch_data)

    # Make predictions for the batch data
    batch_predictions = best_model.predict(batch_data)
    batch_predictions = pd.Series(batch_predictions)
    batch_predictions = batch_predictions.map({
        0: "Low", 1: "Medium", 2: "High"
    })

    # Display the batch predictions
    sd.subheader("Batch Predictions")
    sd.write("Predicted Salary Categories for Batch Data:")
    sd.write(batch_predictions)
    # Save the predictions to the batch data

    batch_data['Predicted Total Salary'] = batch_predictions
    sd.write(batch_data[['Predicted Total Salary']])

sd.write("Batch predictions completed successfully!")
sd.write("Streamlit app is running! Visit the URL to interact with the app.")
sd.write("To run the app, use the command: `streamlit run app.py`")
sd.write("Make sure to have the best_model.pkl file in the same directory as this app.py file.")
```
